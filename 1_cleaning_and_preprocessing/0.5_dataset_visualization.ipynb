```python
from scipy.stats import skew
def calc_skew(dataset):
    return_dict = {}
    for column in independent_vars:
        if column in ['SR_B1','SR_B2','SR_B3','SR_B4','SR_B5','SR_B6','SR_B7']:
            return_dict[column+" point"] = skew(dataset[column+" point"].astype('float64').to_list(), axis = 0, bias = True)
            return_dict[column+" mean"] = skew(dataset[column+" mean"].astype('float64').to_list(), axis = 0, bias = True)
            return_dict[column+" median"] = skew(dataset[column+" median"].astype('float64').to_list(), axis = 0, bias = True)
        else:
            return_dict[column] = skew(dataset[column].astype('float64').to_list(), axis = 0, bias = True)
    return return_dict
```

```python
dataset['month']=dataset['sample_date'].apply(lambda x: x.split('-')[1])
dataset['year']=dataset['sample_date'].apply(lambda x: x.split('-')[0][-2:])
calc_skew(dataset)
```
